## K-Means++

Отличается от обычного K-Means тем, что при инициализации центроидов, вероятность случайного объекта становления центроидов пропорциональна квадрату расстояния от сущетствующих центроидов

Методы расчёта метрик:

````python
EPS = 1e-9

def get_cluster_dists(X: np.ndarray, k_labels, sample_idx, cluster_idx: int):
    """
    Расчёт расстояний от точки x с индексом sample_idx до всех точек кластера с индексом cluster_idx
    """
    if k_labels is None:
        return np.array([0.0])
    
    dists = []

    for sample_j in range(len(X)):
        if k_labels[sample_j] == cluster_idx and sample_j != sample_idx:
            dists.append(np.linalg.norm(X[sample_idx] - X[sample_j]))

    return dists

def silhouette_score(X, k_labels, eps = 1e-6):
    if k_labels is None:
        return 0.0
    
    k_labels = np.asarray(k_labels)
    X = np.asarray(X)

    # -1 игнорируем - модель посчитала выбросом

    mask = (k_labels != -1)
    X = X[mask]
    k_labels = k_labels[mask]

    if len(k_labels) == 0:
        return None

    unique_clusters = np.unique(k_labels)
    if len(unique_clusters) < 2:
        return None
    
    score = 0.0
    valid_samples = 0
    for i_sample in range(len(X)):
        cluster_idx = k_labels[i_sample]

        a_cluster_dists = get_cluster_dists(X, k_labels, i_sample, cluster_idx)
        if len(a_cluster_dists) == 0:
            continue
        a = sum(a_cluster_dists) / len(a_cluster_dists)

        b = np.inf
        for cluster_j in unique_clusters:
            if cluster_idx == cluster_j: continue

            b_cluster_dists = get_cluster_dists(X, k_labels, i_sample, cluster_j)

            if len(b_cluster_dists) == 0: continue
            b = min(b, sum(b_cluster_dists) / len(b_cluster_dists))

        score += silhouette_score_sample(a, b)
        valid_samples += 1

    if valid_samples == 0:
        return 0.0

    return score / valid_samples


def silhouette_score_sample(a, b):
    silhouette_score = (b - a) / max(a, b, EPS)
    return silhouette_score

def homogeneity_score(X, k_labels, c_labels):
    if c_labels is None or k_labels is None:
        return None

    if len(c_labels) == 0 or len(k_labels) == 0:
        return None
    
    k_labels = np.asarray(k_labels, dtype=int)
    c_labels = np.asarray(c_labels, dtype=int)

    mask = (k_labels != -1)
    k_labels = k_labels[mask]
    c_labels = c_labels[mask]

    if len(k_labels) == 0:
        return None

    n = len(k_labels)

    n_classes = max(c_labels) + 1
    n_clusters = max(k_labels) + 1

    m_c = np.bincount(c_labels)
    n_k = np.bincount(k_labels)

    n_ck = np.zeros((n_classes, n_clusters), dtype=int)
    np.add.at(n_ck, (c_labels, k_labels), 1)

    p_c = m_c / n
    h_class = -np.sum(p_c[p_c > 0] * np.log(p_c[p_c > 0]))

    h_class_clust = 0.0
    for k in range(n_clusters):
        if n_k[k] == 0:
            continue
        for c in range(n_classes):
            if n_ck[c, k] == 0:
                continue
            h_class_clust -= (n_ck[c, k] / n) * np.log(n_ck[c, k] / n_k[k])

    if h_class == 0.0:
        return 1.0

    return 1 - h_class_clust / h_class

def completeness_score(X, k_labels, c_labels):
    if c_labels is None or k_labels is None:
        return None

    if len(c_labels) == 0 or len(k_labels) == 0:
        return None
    
    k_labels = np.asarray(k_labels, dtype=int)
    c_labels = np.asarray(c_labels, dtype=int)

    mask = (k_labels != -1)
    k_labels = k_labels[mask]
    c_labels = c_labels[mask]

    if len(k_labels) == 0:
        return None

    n = len(k_labels)

    n_classes = max(c_labels) + 1
    n_clusters = max(k_labels) + 1

    m_c = np.bincount(c_labels)
    n_k = np.bincount(k_labels)

    n_ck = np.zeros((n_classes, n_clusters), dtype=int)
    np.add.at(n_ck, (c_labels, k_labels), 1)

    p_k = n_k / n
    h_clust = -np.sum(p_k[p_k > 0] * np.log(p_k[p_k > 0]))

    h_clust_class = 0.0
    for c in range(n_classes):
        if m_c[c] == 0:
            continue
        for k in range(n_clusters):
            if n_ck[c, k] == 0:
                continue
            h_clust_class -= (n_ck[c, k] / n) * np.log(n_ck[c, k] / m_c[c])
    
    if h_clust == 0.0:
        return 1.0

    return 1 - h_clust_class / h_clust

def v_measure_score(X, k_labels, c_labels, eps=1e-6):
    if c_labels is None or k_labels is None:
        return None

    if len(c_labels) == 0 or len(k_labels) == 0:
        return None
    
    homogeneity = homogeneity_score(X, k_labels, c_labels)
    completeness = completeness_score(X, k_labels, c_labels)

    if homogeneity is None or completeness is None:
        return None
    
    v_score = 2 * homogeneity * completeness / (homogeneity + completeness + eps)
    return v_score
````

````python
class KMeansPP:
    def __init__(self,
            k=3,
            max_iter = 1000,
            random_seed = 42,
            eps = 1e-6,
            ):
        self.k = k
        self.max_iter = max_iter
        self.random_seed = random_seed
        self.rng = np.random.default_rng(self.random_seed)
        self.eps = eps

        self.labels_ = None
        self.metrics_ = {
            'silhouette': 0.0,
            'homogeneity': 0.0,
            'completeness': 0.0,
            'v1-score': 0.0
        }

        if self.max_iter <= 0:
            raise ValueError("max_iter > 0")

    def init_centroids_pp(self, X: np.ndarray):
        """
        Инициализация центроидов для алгоритма KMeans++
        Вероятность выбора объекта как центроида пропорциональна квадрату расстояния от него до ближайшего уже выбранного центроида
        """
        self.centroids = np.full((self.k, X.shape[1]), np.nan)
        centroid_idx = self.rng.choice(len(X))
        self.centroids[0] = X[centroid_idx]

        for i in range(1, self.k):
            # расчёт дистанций от каждой точки до ближайшего центроида
            distances, _ = self.get_distances(X)
            distances = np.pow(distances, 2)
            p = distances / distances.sum()
            centroid_idx = self.rng.choice(len(X), p=p)
            self.centroids[i] = X[centroid_idx]

    def get_distances(self, X: np.ndarray):
        """
        Расчёт расстояний от каждой точки до ближайшего центроида
        Возвращаются дистанции, и индексы ближайшего центроида для каждого объекта
        """
        distances = np.full(len(X), np.inf)
        nearest_centroids = np.full(X.shape[0], -1, dtype=int)

        for i, centroid in enumerate(self.centroids):
            if np.isnan(centroid).any(): break
            # X : [n,m]
            # X[centroid] : [m]
            dist = np.linalg.norm(X - centroid, axis=1)

            mask = dist < distances
            distances[mask] = dist[mask]
            nearest_centroids[mask] = i
            
        return distances, nearest_centroids

    
    def update_centroids(self, X):

        for i in range(self.k):
            X_centroid = X[self.labels_ == i]

            if len(X_centroid) > 0:
                self.centroids[i] = X_centroid.mean(axis=0)

    def fit(self, X: np.ndarray):
        self.init_centroids_pp(X)
        self.labels_ = np.full(len(X), -1, dtype=int)

        last_inertia = np.inf
        for _ in range(self.max_iter):
            min_dist, nearest_centroids = self.get_distances(X)

            self.inertia_ = np.sum(min_dist**2)
            self.labels_ = nearest_centroids

            self.update_centroids(X)

            if abs(self.inertia_ - last_inertia) < self.eps:
                break

            last_inertia = self.inertia_

        
        min_dist, nearest_centroids = self.get_distances(X)
        self.inertia_ = np.sum(min_dist**2)
        self.labels_ = nearest_centroids

        return self

    def count_metrics(self, X, c_labels=None):
        if c_labels is None:
            homogeneity = 0.0
            completeness = 0.0
            v_score = 0.0
        else:
            homogeneity = homogeneity_score(X, self.labels_, c_labels)
            completeness = completeness_score(X, self.labels_, c_labels)
            v_score = v_measure_score(X, self.labels_, c_labels, self.eps)
            
        silhouette = silhouette_score(X, self.labels_)

        self.metrics_ = {
            'silhouette': silhouette,
            'homogeneity': homogeneity,
            'completeness': completeness,
            'v1-score': v_score
        }

        return self.metrics_
````

````python
model = KMeansPP(k=2)
model.fit(X_scaled)
metrics = model.count_metrics(X_scaled, c_labels)
for metric, score in metrics.items():
    print(f"{metric}: \t{round(score,3)}")
````

# DBSCAN

````python
from collections import deque
````

````python
class DBSCAN:
    def __init__(
        self,
        min_samples=3,
        eps = 0.5,
    ):
        self.min_samples = min_samples
        self.eps = eps

        self.labels_ = None
        self.neighbors_cache_ = {}
        self.metrics_ = {
            'silhouette': 0.0,
            'homogeneity': 0.0,
            'completeness': 0.0,
            'v1-score': 0.0
        }

    def get_ball_samples(self, X, sample_idx) -> list[int]:
        """
        Получить все индексы точек в шаре B(eps, X[sample_idx])
        """
        ball_samples = []

        if sample_idx in self.neighbors_cache_:
            return self.neighbors_cache_[sample_idx]

        for idx in range(len(X)):
            dist = np.linalg.norm(X[sample_idx] - X[idx])

            if dist <= self.eps:
                ball_samples.append(idx)

        self.neighbors_cache_[sample_idx] = ball_samples

        return ball_samples
    
    def classify_samples(self, X):
        # основные точки: в окрестности не менее чем self.min_samples точек
        core_samples_idx = set()

        for sample_idx in range(len(X)):
            neighbors = self.get_ball_samples(X, sample_idx)

            if len(neighbors) >= self.min_samples:
                core_samples_idx.add(sample_idx)

        border_samples_idx = []
        outliers_idx = []

        # граничные/шумовые
        for sample_idx in range(len(X)):
            if sample_idx in core_samples_idx: continue
            neighbors = self.get_ball_samples(X, sample_idx)

            is_border = 0
            for neighbor_idx in neighbors:
                if neighbor_idx in core_samples_idx:
                    border_samples_idx.append(sample_idx)
                    is_border = 1
                    break
            if not is_border:
                outliers_idx.append(sample_idx)
        
        return sorted(core_samples_idx), border_samples_idx, outliers_idx
    
    def extend_cluster(self, X, core_idx, visited):
        if self.labels_ is None:
            raise ValueError("model not fitted: labels_ is None")

        cur_label = self.labels_[core_idx]

        # очередь, объекты которой закидываем в кластер текущей core точки
        neighbors = deque()
        neighbors.extend(self.get_ball_samples(X, core_idx))

        while neighbors:
            neighbor_idx = neighbors.popleft()

            if neighbor_idx in visited: continue

            # если сосед тоже core, добавляем его соседей в очередь
            if neighbor_idx in self.core_set:
                self.labels_[neighbor_idx] = cur_label
                visited.add(neighbor_idx)

                neighbors.extend(self.get_ball_samples(X, neighbor_idx))

            # если сосед border, то добавляем в кластер, но его соседей игнорируем
            else:
                self.labels_[neighbor_idx] = cur_label
                visited.add(neighbor_idx)

    def fit(self, X):

        self.neighbors_cache_ = {}
        self.labels_ = np.full(len(X), -1, dtype=int)
        
        self.n_clusters_ = 0
        visited = set()

        self.core_samples_idx, _, _ = self.classify_samples(X)
        self.core_set = set(self.core_samples_idx)

        for core_idx in self.core_samples_idx:

            # если точка core и еще не размечена - добавляем в новый кластер
            if core_idx not in visited:
                self.labels_[core_idx] = self.n_clusters_
                self.n_clusters_ += 1
            else:
                continue

            visited.add(core_idx)
            

            self.extend_cluster(X, core_idx, visited)

    def count_metrics(self, X, c_labels=None):

        if c_labels is None or self.labels_ is None:
            homogeneity = 0.0
            completeness = 0.0
            v_score = 0.0
        else:
            # mask = (self.labels_ != -1)
            # X_f = X[mask]
            # k_labels_f = self.labels_[mask]
            # c_labels_f = c_labels[mask]

            # homogeneity = homogeneity_score(X_f, k_labels_f, c_labels_f)
            # completeness = completeness_score(X_f, k_labels_f, c_labels_f)
            # v_score = v_measure_score(X_f, k_labels_f, c_labels_f, self.eps)

            homogeneity = homogeneity_score(X, self.labels_, c_labels)
            completeness = completeness_score(X, self.labels_, c_labels)
            v_score = v_measure_score(X, self.labels_, c_labels)
            
        silhouette = silhouette_score(X, self.labels_)

        self.metrics_ = {
            'silhouette': silhouette,
            'homogeneity': homogeneity,
            'completeness': completeness,
            'v1-score': v_score
        }

        return self.metrics_
````
